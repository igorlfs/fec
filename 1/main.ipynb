```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 1000
```

# Exercício 1

```python
def func(r: float, y0: float):
    y = [y0]
    for i in range(N - 1):
        y.append(r * y[i] * (1 - y[i]))
    x = range(N)
    plt.xlabel("Geração")
    plt.ylabel("População")
    plt.title(f"r = {r} $y_0$ = {y0}")
    plt.plot(x, y)


func(0.5, 0.5)
```

# Exercício 2

```python
rs = [2.5, 3.1, 3.5, 3.7]
for r in rs:
    func(r, 0.5)
    plt.show()
```

# Exercício 3

```python
y0s = [0.25, 0.5, 0.75]
for r in rs:
    for y0 in y0s:
        func(r, y0)
    plt.show()
```

# Exercício 4

```python
y0s = [0.5, 0.501, 0.5001]
for r in rs:
    for y0 in y0s:
        func(r, y0)
    plt.show()
```

# Exercício 5

```python
r_linspace = np.linspace(10e-5, 4, 10000)
Y: list[float] = list()
R: list[float] = list()
for r in r_linspace:
    R.append(r)
    y = np.random.random()
    for _ in range(N - 1):
        y = (r * y) * (1 - y)
    Y.append(y)
plt.plot(R, Y, ls="", marker=",")
plt.ylabel("$y_0$")
plt.xlabel("$r$")
plt.show()
```
